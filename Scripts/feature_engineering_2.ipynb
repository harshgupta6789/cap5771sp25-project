```python
# Sample: df['CRSDepTime'] is in HHMM format like 1410, 1604
def convert_to_hour(time_val):
    """Convert HHMM int to hour in 24h format."""
    hour = time_val // 100
    return hour % 24  # modulo 24 just in case of weird data

# Apply to create hour columns
df['DepHour'] = df['CRSDepTime'].apply(convert_to_hour)
df['ArrHour'] = df['CRSArrTime'].apply(convert_to_hour)

# Function to map hours to time-of-day bins
def get_time_of_day(hour):
    if 5 <= hour <= 11:
        return 'morning'
    elif 12 <= hour <= 16:
        return 'afternoon'
    elif 17 <= hour <= 20:
        return 'evening'
    else:
        return 'night'

# Create new binned features
df['DepTimeOfDay'] = df['DepHour'].apply(get_time_of_day)
df['ArrTimeOfDay'] = df['ArrHour'].apply(get_time_of_day)
```

```python
df['ArrDelay']
```

```python
df['Distance']
```

```python
def distance_bin(miles):
    if miles <= 500:
        return 'short'
    elif miles <= 1500:
        return 'medium'
    else:
        return 'long'

df['DistanceBin'] = df['Distance'].apply(distance_bin)
```

```python
df['DepDelay_Short'] = df.apply(lambda x: x['DepDelay'] if x['DistanceBin'] == 'short' else 0, axis=1)
df['DepDelay_Medium'] = df.apply(lambda x: x['DepDelay'] if x['DistanceBin'] == 'medium' else 0, axis=1)
df['DepDelay_Long'] = df.apply(lambda x: x['DepDelay'] if x['DistanceBin'] == 'long' else 0, axis=1)
```

Interaction term

```python
df['DepDelay_Distance_Interaction'] = df['DepDelay'] * df['Distance']
```

```python
df['Route'] = df['OriginState'] + '-' + df['DestState']
```

```python
# Average departure delay per airline
avg_dep_delay_airline = df.groupby('Reporting_Airline')['DepDelay'].mean().reset_index(name='AvgDepDelay')

# Average arrival delay per airline
avg_arr_delay_airline = df.groupby('Reporting_Airline')['ArrDelay'].mean().reset_index(name='AvgArrDelay')
```

```python
# Average departure delay per route
avg_dep_delay_route = df.groupby('Route')['DepDelay'].mean().reset_index(name='AvgDepDelay')

# Average arrival delay per route
avg_arr_delay_route = df.groupby('Route')['ArrDelay'].mean().reset_index(name='AvgArrDelay')
```

```python
avg_dep_delay_route
```

```python
avg_arr_delay_route
```

```python
avg_dep_delay_airline.rename(columns={'AvgDepDelay': 'AvgDepDelayByAirline'}, inplace=True)
avg_dep_delay_route.rename(columns={'AvgDepDelay': 'AvgDepDelayByRoute'}, inplace=True)
```

```python
# Merge airline average back to original DataFrame
df = df.merge(avg_dep_delay_airline, on='Reporting_Airline', how='left')

# Merge route average back to original DataFrame
df = df.merge(avg_dep_delay_route, on='Route', how='left')
```

```python
df['RouteTraffic'] = df.groupby(['Route', 'FlightDate'])['FlightDate'].transform('count')
```

```python
df['FlightDate'] = pd.to_datetime(df['FlightDate'], format='%Y-%m-%d')
```

```python
df['DayOfWeek'] = df['FlightDate'].dt.dayofweek
```

```python
weather_agg = df.groupby(['DestState', 'FlightDate']).agg(
    WeatherEventCount=('EVENT_TYPE', 'count'),
    # Flag severe weather: 1 if any event on that day in that state had MAGNITUDE > 5, else 0.
    SevereWeatherFlag=('MAGNITUDE', lambda x: 1 if (x > 5).any() else 0)
).reset_index()
```

```python
weather_agg.rename(columns={'WeatherEventCount': 'WeatherEventCountByState'}, inplace=True)
```

```python
df = df.merge(weather_agg[['DestState', 'FlightDate', 'WeatherEventCount', 'SevereWeatherFlag']],
              left_on=['DestState', 'FlightDate'],
              right_on=['DestState', 'FlightDate'],
              how='left')
```
