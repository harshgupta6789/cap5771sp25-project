```python
# Visualisations

# Monthly Passengers Over Time (Line Plot)
xticks_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
monthly_passengers = df_passengers.groupby('Month')['Total'].sum()/1000000
de.create_line_plot(monthly_passengers, title='Monthly Passengers', xlabel='Month',
                    ylabel='Total Passengers (in millions)',
                    xticks=range(1, 13), xtickslabels=xticks_labels)

# Passenger numbers peak during summer (July-August), indicating a seasonal travel trend. The lowest numbers are in February and November.
```

```python
# Bar Chart of Traffic by Airline (Carrier)

airline_traffic = df_passengers.groupby('carrier')['Total'].sum().sort_values(ascending=False)

# Get top 10 airlines
top_airlines = airline_traffic.head(10)/1000000

de.create_bar_chart(top_airlines, title='Top 10 Airlines by Total Passengers',
                    xlabel='Airline', ylabel='Total Passengers (in millions)')
# Airline 'AA' dominates in total passengers, followed by 'UA' and 'DL', suggesting market concentration among these carriers
```

```python
# Total Flights per Year

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Scheduled', data=df_departures, label='Scheduled Flights')
sns.lineplot(x='Year', y='Charter', data=df_departures, label='Chartered Flights')
sns.color_palette("flare", as_cmap=True)
plt.title('Total Flights per Year')
plt.xlabel('Year')
plt.ylabel('Total Flights (in Million)')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.show()

# Scheduled flights show consistent growth until a sharp decline post-2020, likely due to COVID-19. Chartered flights remain negligible throughout the years.
```
